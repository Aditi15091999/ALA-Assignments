```python
import numpy as np
import matplotlib.pyplot as plt

a = np.arange(start=0, stop=5, step=0.1)
b = np.arange(start=0, stop=5, step=0.1)
eqn= 3*a**2+2*a*b 

#solving for 2 and 3
z= 18*a+4*b-24
fig = plt.figure()

ax = fig.add_subplot(111, projection='3d')
ax.plot(a,b,eqn)
ax.plot(a,b,z)
plt.show()
```
